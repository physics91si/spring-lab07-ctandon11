## Part 1

```python
# Load image here. Display the image and the image shape.
image=io.imread('starry.jpg')
io.imshow(image)
print(np.shape(image))
```

```python
# Make the color arrays and display them on a labelled subplot.
red=np.copy(image)
red[:,:,[1,2]]=0
blue=np.copy(image)
blue[:,:,[0,2]]=0
green=np.copy(image)
green[:,:,[0,1]]=0
f, images = plt.subplots(1,3)
images[0].imshow(red)
images[1].imshow(blue)
images[2].imshow(green)
images[0].set_title('red')
images[1].set_title('green')
images[2].set_title('blue')
```

## Part 2

```python
from math import sqrt
from skimage.feature import blob_log, blob_doh
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Here is the image of the Hubble Deep Field. 
image = data.hubble_deep_field()[0:500, 0:500]
gray=rgb2gray(image)
f, images = plt.subplots(1,2)
images[0].imshow(image)
images[1].imshow(gray,cmap='gray')
images[0].set_title('original')
images[1].set_title('black and white')

#Convert the image to grayscale and display next to the original.
```

```python
# Apply the two Blob detection methods and measure the time taken by the algorithms.
blobs_log=blob_log(gray, min_sigma=1,max_sigma=50, threshold=0.2)
blobs_doh=blob_doh(gray, max_sigma=50, threshold=0.01)
print(blobs_log)
print(blobs_doh)
```

```python
# Plot the 2 methods side-by-side along with circle patches.
doh=np.copy(gray)
log=np.copy(gray)
fig, ax=plt.subplots(1,2)
for index in blobs_doh:
    p=plt.Circle((index[1],index[0]),radius=index[2],fill=False, edgecolor="w")
    ax[0].imshow(doh)
    ax[0].add_patch(p)
for index in blobs_log:
    p=plt.Circle((index[1],index[0]),radius=index[2],fill=False, edgecolor="w")
    ax[1].imshow(log)
    ax[1].add_patch(p)
```
